```python
df1 = pd.read_csv('/content/drive/My Drive/case_study/202101-divvy-tripdata.csv')
df2 = pd.read_csv('/content/drive/My Drive/case_study/202102-divvy-tripdata.csv')
df3 = pd.read_csv('/content/drive/My Drive/case_study/202103-divvy-tripdata.csv')
df4 = pd.read_csv('/content/drive/My Drive/case_study/202104-divvy-tripdata.csv')
df5 = pd.read_csv('/content/drive/My Drive/case_study/202105-divvy-tripdata.csv')
df6 = pd.read_csv('/content/drive/My Drive/case_study/202106-divvy-tripdata.csv')
df7 = pd.read_csv('/content/drive/My Drive/case_study/202107-divvy-tripdata.csv')
df8 = pd.read_csv('/content/drive/My Drive/case_study/202108-divvy-tripdata.csv')
df9 = pd.read_csv('/content/drive/My Drive/case_study/202109-divvy-tripdata.csv')
df10 = pd.read_csv('/content/drive/My Drive/case_study/202110-divvy-tripdata.csv')
df11 = pd.read_csv('/content/drive/My Drive/case_study/202111-divvy-tripdata.csv')
df12 = pd.read_csv('/content/drive/My Drive/case_study/202112-divvy-tripdata.csv')
```

```python
#Merging the whole 12 monthly csv files 
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10])
df_final_concat = pd.concat([df, df11, df12])
df_final_concat
```

```python
df_final_concat.describe()
```

```python
df_final_concat.info()
```

```python
#Changing the datatypes for the time which the riders started riding and the end time 
df_final_concat['started_at'] = pd.to_datetime(df_final_concat['started_at'])
df_final_concat['ended_at'] = pd.to_datetime(df_final_concat['ended_at'])
```

```python
df_final_concat.info()
```

```python
df_member = df_final_concat.loc[df_final_concat['member_casual'] == "member"]
df_member
```

```python
df_casual = df_final_concat.loc[df_final_concat['member_casual'] == "casual"]
df_casual
```

from the two dataframes we can tell that there are more members than casual buyers

```python
df_final_concat["Duration"] =  (df_final_concat['ended_at']-df_final_concat['started_at']).astype('timedelta64[m]')
```

```python
df_final_concat.head()
```

```python
#Filtering data for the riders who rode the bicycles for less than a minute
df_duration_filter = df_final_concat.loc[df_final_concat['Duration'] == 0]
df_duration_filter
```

```python
#We want to find and create a column on which day of the week did the riders ride their bicycles which day of the week 
#df_final_concat['day_of_the_week'] = df_final_concat['started_at'].dt.day_name()
df_final_concat['day_of_the_week'] = df_final_concat['started_at'].dt.weekday
```

```python
df_final_concat
```

```python
#Here we want to find aggregate values for the mean, max and mode of the duration and day of the week respectively
#the duration is with respect to  hours
mode_value = df_final_concat['day_of_the_week'].mode()
mean_duration = df_final_concat['Duration'].mean()
max_duration = df_final_concat['Duration'].max()
print(mode_value)
print(mean_duration)
print(max_duration)
```

```python
#Filtering data for the riders who rode the bicycles for less than a minute
df_duration_filters = df_final_concat.loc[df_final_concat['Duration']  < 60]
df_duration_filters
```
